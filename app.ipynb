```python
import pandas as pd
```

```python
df1 = pd.read_csv("sales_january_2019.csv")
df2 = pd.read_csv("Sales_February_2019.csv")
df3 = pd.read_csv("Sales_March_2019.csv")
df4 = pd.read_csv("Sales_April_2019.csv")
df5 = pd.read_csv("Sales_May_2019.csv")
df6 = pd.read_csv("Sales_June_2019.csv")
df7 = pd.read_csv("Sales_July_2019.csv")
df8 = pd.read_csv("Sales_August_2019.csv")
df9 = pd.read_csv("Sales_September_2019.csv")
df10 = pd.read_csv("Sales_November_2019.csv")
df11 = pd.read_csv("Sales_October_2019.csv")
df12 = pd.read_csv("Sales_December_2019.csv")

```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12],ignore_index=True)
```

```python
df = df.dropna()
```

```python
df
```

```python
df.to_csv("annual_data.csv",index=False)
```

```python
df = df.set_index("Order ID")
```

```python
df['Order Date']=pd.to_datetime(df["Order Date"],errors="coerce")
```

```python
df.info()
```

```python
df['Month']=df['Order Date'].dt.month_name()
```

```python
df
```

```python
!pip install matplotlib
import matplotlib
order_count = df.groupby("Month").size()
```

```python
import matplotlib.pyplot as plt
order_count.plot(kind='pie',autopct='%1.1f%%')
plt.title('Numbers of orders per month')
plt.show()
```
